###### Collect Team Performance Data

###### Launch Web Browser (Chrome) using Selenium and launch the URL for player statistics

###### Generate Statistics for Training Data (Seasons 1 to 7)

```python
driver = webdriver.Chrome()
driver.get('https://www.prokabaddi.com/stats/49-96-total-points-statistics')
time.sleep(5)

season_list = ['Season 1', 'Season 2', 'Season 3', 'Season 4', 'Season 5', 'Season 6', 'Season 7']
stats_list = ['Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded']

train_df = pd.DataFrame(columns = ['Season', 'Team', 'Matches Played', 'Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded'])

for season in season_list:
    season_element = driver.find_element_by_id("season_dropdown")
    season_element.send_keys(season)
    time.sleep(3)
    
    element = driver.find_element_by_id('load_more')
    while (driver.find_element_by_id('load_more')):
        try:
            element.click()
            sleep(3)
        except Exception as e:
            break
    
    soup = BeautifulSoup(driver.page_source)
    
    y_team = []
    y_team = soup.findAll('div',attrs={'class':'sipk-lb-playerName'})
    y_team_list=list()
    for j in range(0, len(y_team)):
            y_team_list.append(y_team[j].span.string.strip())

    y_matches_played = []
    y_matches_played = soup.findAll('div',attrs={'class':'sipk-lb-matchedPlayed'})
    y_matches_played_list=list()
    season_list = list()
    for j in range(0, len(y_matches_played)):
            y_matches_played_list.append(y_matches_played[j].span.string.strip())
            season_list.append(season)

    temp_df_1 = pd.DataFrame(columns = ['Season', 'Team', 'Matches Played', 'Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded'])
    temp_df_1['Season'] = season_list
    temp_df_1['Team'] = y_team_list
    temp_df_1['Matches Played'] = y_matches_played_list

    
    
    #temp_df = pd.DataFrame(columns = ['Team', 'Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded'])
    for stats in stats_list:
        stats_element1 = driver.find_element_by_id("si_dropdown")
        stats_element1.send_keys(stats)
        time.sleep(3)
        
        element1 = driver.find_element_by_id('load_more')
        while (driver.find_element_by_id('load_more')):
            try:
                element1.click()
                time.sleep(3)
            except Exception as e:
                break
        
        soup_1 = BeautifulSoup(driver.page_source)
        
        x_points=[]
        x_points = soup_1.findAll('div',attrs={'class':'sipk-lb-raidPoints'})
        points_list = list()        
        for i in range(0, len(x_points)):
            points_list.append(x_points[i].span.string.strip())
        
        x_team=[]
        x_team = soup_1.findAll('div',attrs={'class':'sipk-lb-playerName'})
        team_list=list()
        for i in range(0, len(x_team)):
            team_list.append(x_team[i].span.string.strip())
            
        
        temp_df = pd.DataFrame(columns = ['Team', stats])
        temp_df['Team'] = team_list
        temp_df[stats] = points_list
        
        temp_df_1 = pd.merge(temp_df, temp_df_1, on = 'Team', how="inner")
        temp_df_1 = temp_df_1.rename(columns = {"All-outs conceded_x" : "All-outs conceded", "All-outs inflicted_x" : "All-outs inflicted", "Do-or-Die raid points_x" : "Do-or-Die raid points", "Super tackles_x" : "Super tackles", "Super raids_x" : "Super raids", "Avg tackle points_x" : "Avg tackle points", "Tackle points_x" : "Tackle points", "Successful tackles_x" : "Successful tackles", "Avg raid points_x" : "Avg raid points", "Raid points_x" : "Raid points", "Successful raids_x" : "Successful raids", "Avg points scored_x" : "Avg points scored", "Total points conceded_x" : "Total points conceded", "Total points scored_x" : "Total points scored"})
        temp_df_1 = temp_df_1[['Season', 'Matches Played', 'Team', 'Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded']]
            
    train_df = train_df.append(temp_df_1)
train_df = train_df[['Season', 'Team', 'Matches Played', 'Total points scored', 'Total points conceded', 'Avg points scored', 'Successful raids', 'Raid points', 'Avg raid points', 'Successful tackles', 'Tackle points', 'Avg tackle points', 'Super raids', 'Super tackles', 'Do-or-Die raid points', 'All-outs inflicted', 'All-outs conceded']]
train_df = train_df.reset_index(drop = True)
driver.quit()
del driver
```
